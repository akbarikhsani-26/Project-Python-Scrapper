```python
import csv
import re
import requests
from bs4 import BeautifulSoup

url = "https://realpython.github.io/fake-jobs/"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

quotes = soup.find_all('div', class_='card-content')

all_jobs_data = []

for q in quotes:
    job_el = q.find('h2', class_='title is-5')
    job = job_el.text.strip() if job_el else "N/A"

    company_el = q.find('h3', class_='company')
    company = company_el.text.strip() if company_el else "N/A"

    location_el = q.find('p', class_='location')
    location = location_el.text.strip() if location_el else "N/A"

    link_el = q.find('a', class_='card-footer-item', string=re.compile(r'Apply', re.IGNORECASE))
    link_url = link_el['href'].strip() if (link_el and 'href' in link_el.attrs) else "Link Tidak Tersedia"

    if job != "N/A" or company != "N/A":
        all_jobs_data.append({
            'Job Title': job,
            'Company': company,
            'Location': location,
            'Apply Link': link_url
        })

        print(f"Job: {job}\nCompany: {company}\nLocation: {location}\nLink: {link_url}\n{'-'*30}")
```

```python
nama_file = 'Scrapping_Python.csv'

with open(nama_file, mode='w', newline='', encoding='utf-8') as file:
    fieldnames = ['Job Title', 'Company', 'Location', 'Apply Link']
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()
    writer.writerows(all_jobs_data)

print(f"\n🔥 Selesai! {len(all_jobs_data)} data berhasil disimpan di '{nama_file}'")
```
